```python
#Object Detection
import cv2
import numpy as np

while True:
    img = cv2.imread('img_path')

    #range of colors
    lower = np.array(['put color range'])
    upper = np.array(['put color range'])

    mask = cv2.inRange(img, lower, upper)
    result = cv2.bitwise_and(img, img, mask = mask)
    if cv2.waitKey(0) & 0xFF == ord('e'):
        break

    cv2.imshow('Original', img)
    cv2.imshow('Detect', result)

cv2.destroyAllWindows()
```

```python
# Face Detection Picture
import cv2

#Read
img = cv2.imread('img_path')

#read classifier
face_cascade = cv2.CascadeClassifier('path/haarcascade_frontalface_default.xml')

#convert bgr img to gray img
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#face detection process
face_detect = face_cascade.detectMultiScale(gray_img, 1.1, 3)

#draw rectangle on face
for (x,y,w,h) in face_detect:
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), thickness = 5)

#show img
cv2.imshow('title', img)
cv2.waitKey(0)
cv2.destroyAllWindows()


```

```python
#Face Detection Video
import cv2

#read video
cap = cv2.VideoCapture('video_path')

#read classifier
face_cascade = cv2.CascadeClassifier('path/haarcascade_frontalface_default.xml')

#show Video window
while (cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        face_detect = face_cascade.detectMultiScale(gray_img, 1.1, 3)
        for (x,y,w,h) in face_detect:
            cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0), thickness = 5)
            cv2.imshow('title', frame)
        if cv2.waitKey(1) & 0xFF == ord('e'):
            break
        else:
            break
cap.release()
cv2.destroyAllWindows()
```

```python
#Face + Eyes Detection Picture
import cv2

#read img
img = cv2.imread('img_path')

#read xml
face_cascade = cv2.CascadeClassifier('path/haarcascade_frontalface_default.xml')
eyes_cascade = cv2.CascadeClassifier('path/haarcascade_eye_tree_eyeglasses.xml')

#rbg to gray
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

#face detect
face_detect = face_cascade.detectMultiScale(gray_img, 1.1, 3)

#eyes detect
eyes_detect = eyes_cascade.detectMultiScale(gray_img, 1.1, 3)

for (fx,fy,fw,fh) in face_detect:
    cv2.rectangle(img, (fx,fy), (fx+fw, fy+fh), (255,0,0), thickness = 5)
    for (ex,ey,ew,eh) in eyes_detect:
        cv2.rectangle(img, (ex,ey), (ex+ew, ey+eh), (255,0,0), thickness = 5)

cv2.imshow('title', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Motion Detection (Contour)
import cv2

cap = cv2.VideoCapture('Video_path')
ret, frame1 = cap.read()
ret, frame2 = cap.read()

while cap.isOpened():
    if ret == True:
        motion_diff = cv2.absdiff(frame1, frame2)
        gray_img = cv2.cvtColor(motion_diff, cv2.COLOR_BGR2GRAY)

        blur = cv2.GaussianBlur(gray_img, (5,5), 0)
        thresh, result = cv2.threshold(blur, 15, 255, cv2.THRESH_BINARY)
        dilation = cv2.dilate(result, None, iteration = 3)

        contours, hierarchy = cv2.findContours(dilation, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
        cv2.drawContours(frame1, contours, -1, (255, 0, 0), 2)
        cv2.imshow('title', frame1)
        frame1 = frame2
        ret, frame2 = cap.read()
        if cv2.waitKey(1) & 0xFF == ord('e'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
#Montion Detection Marker
import cv2

cap = cv2.VideoCapture('Video_path')
ret, frame1 = cap.read()
ret, frame2 = cap.read()

while cap.isOpened():
    if ret == True:
        motion_diff = cv2.absdiff(frame1, frame2)
        gray_img = cv2.cvtColor(motion_diff, cv2.COLOR_BGR2GRAY)

        blur = cv2.GaussianBlur(gray_img, (5,5), 0)
        thresh, result = cv2.threshold(blur, 15, 255, cv2.THRESH_BINARY)
        dilation = cv2.dilate(result, None, iteration = 3)

        contours, hierarchy = cv2.findContours(dilation, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)

        for contour in contours:
            x,y,w,h = cv2.boundingRect(contour)
            if cv2.contourArea(contour) < 2000:
                continue
            cv2.rectangle(frame1, (x,y), (x+w, y+h), (255,0,0), 2)

        cv2.imshow('title', frame1)
        frame1 = frame2
        ret, frame2 = cap.read()
        if cv2.waitKey(1) & 0xFF == ord('e'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
